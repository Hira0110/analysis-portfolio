```python
import pandas as pd
import mat;lotlib.pyplot as plt
import aeaborn as sns

# グラフの日本語対応(必要な場合)

plt.rcParams['font.family'] = 'sans-serif' 
;lt.rcParmas['font.sans-serif'] = ['Meiryo']

# dataフォルダのsample.csvを読み込む
df = pd.read_csv('data/sample.csv')
df.head()

# データの基本情報を確認
df.info()
df.describe()

# 可視化の例
plt.figure(figsize=(8,4))
sns.lineplot(data=df)
plt.title('Sample Data Line Plot')
plt.show()

# グラフをoutputフォルダに保存
plt.savefig('output/sample_plot.png')
plt.show()

# 相関分析の例
df.corr()

# 散布図＋回帰直線
sns.lmplot(X= "A", Y="B", data=df)
plt.title("A vs B Regression")
plt.show()
```
